데이콘에서 주최한 전력사용량 예측 AI개발 대회에 참여한 경험을 바탕으로 제주도의 기상정보와 전력수요량의 관계를 찾아보고 예측하는 모델을 만든다.

제주도의 기상정보는 기상청이 기상자료개방포털에서 제공하는 방재기상관측(AWS)자료에서 제주도의 기온, 바람, 강수량, 습도, 기압 데이터를 사용한다.
제주도의 전력수요량 정보는 한국전력거래소에서 공공데이터 포털에 제공하는 시간별 제주 전력수요량 데이터를 사용한다.

## Import

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
#import seaborn as sns

matplotlib.rcParams['font.family'] ='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] =False
```

## Load

```python
df_weather = pd.read_csv('방재기상관측(AWS) 제주.csv', encoding='cp949')
df_power = pd.read_csv('한국전력거래소 시간별 제주 전력수요량.csv', encoding='cp949')
```

```python
df_weather
```

```python
df_power
```

## 데이터 전처리

### 기상관측정보

```python
df_weather.info()
```

Dtype에서 숫자 데이터는 모두 float64와 int64인 것을 보아 형식이 일치하지 않는 데이터는 없을 것을 예측할 수 있다. 다만 데이터들의 결측치가 다량 존재하고 기압과 같은 경우는 절반 가량이 결측치인 것을 확인할 수 있다.

우선 결측치가 많은 현지기압(hPa)와 해면기압(hPa) 데이터를 제거한다.

```python
df_weather.drop(['현지기압(hPa)', '해면기압(hPa)'], axis=1, inplace=True)
```

강수량의 경우 비가 오지 않았을 때 결측되었을 것으로 예상하고 결측치는 0으로 대체한다.

```python
df_weather['강수량(mm)'].fillna(0, inplace=True)
```

나머지 풍향, 풍속, 기온, 습도은 연속된 시간대와 비슷하기에 결측치의 앞의 값으로 대체한다.

```python
df_weather.fillna(method='ffill', inplace=True)
```

```python
df_weather.info()
```

모든 결측치가 채워젔음을 알 수 있다.

우리가 가진 전력수요량의 데이터는 제주 전역의 시간별 데이터이므로 지점과 지점명을 제하고 일시별로 평균을 내준다.

```python
df_weather.drop(['지점', '지점명'], axis=1, inplace=True)
```

```python
df = df_weather.groupby('일시').mean()
```

```python
df
```

일시 정보를 월, 일, 시간으로 바꾼다.

```python
df['일시'] = df.index
df['월'] = df['일시'].apply(lambda x: int(x[5:7]))
df['일'] = df['일시'].apply(lambda x: int(x[8:10]))
df['시간'] = df['일시'].apply(lambda x: int(x[11:13]))
```

```python
df.drop('일시', axis=1, inplace=True)

df
```

### 전력수요량
기상정보는 2021-01-01 0시부터 2021-12-13 23시까지의 데이터를 제공하고 전력수요량은 2007-01-01 1시부터 2021-12-13 24시 까지의 데이터를 제공하기에 전력수요량도 기상정보와 같은 2021-01-01 0시부터 2021-12-13 23시 까지의 데이터를 사용하며 날짜별로 1시에서 24시가 아닌 0시에서 23시의 데이터로 바꾸어 사용하여 기존 데이터프레임에 더한다.

```python
df_power[df_power['거래일자']=='2021-01-01'].index
```

```python
df_power[df_power['거래일자']=='2021-12-31'].index
```

```python
power_list = []
for i in range(5114, 5479):
    power_list += df_power.iloc[i].to_list()[1:]
```

12-31 24시 데이터를 제거한다.

```python
power_list = power_list[:-1]
```

```python
df['전력수요량(MWh)'] = power_list
```

```python
df
```

```python
df.reset_index(inplace=True)
```

```python
df.drop('일시', axis=1, inplace=True)
```

```python
df.describe()
```

## 데이터 시각화

### 기상 데이터와 전력수요량의 연관성

```python
df.groupby(df['기온(°C)'].apply(lambda x:(x//0.5)/2)).mean()['전력수요량(MWh)'].plot()
```

기온이 15~20에 가까울 수록 전력수요량은 낮고 그에 멀어질 수록 증가함을 확인할 수 있다.

```python
df.groupby(df['강수량(mm)'].apply(lambda x:(x//0.5)/2)).mean()['전력수요량(MWh)'].plot()
```

명확한 관계를 확인하기 어렵다.

```python
df.groupby(df['습도(%)'].apply(lambda x:(x//1))).mean()['전력수요량(MWh)'].plot()
```

습도가 높을수록 전력수요량이 증가함을 확인할 수 있다. 하지만 scatter로 확인해 보면

```python
plt.scatter(df['습도(%)'],df['전력수요량(MWh)'], s=0.5)
plt.xlabel('습도(%)')
plt.ylabel('전력수요량(MWh)')
```

명확한 상관관계를 가진다고 말하기엔 부족하다.

```python
df.groupby(df['풍속(m/s)'].apply(lambda x:(x//0.5)/2)).mean()['전력수요량(MWh)'].plot()
```

```python
df['풍속(m/s)'].describe()
```

대부분이 몰려있는 2~4 구간에서 뚜렷한 상관관계를 나타내지 못한다.

```python
df.groupby(df['풍향(deg)'].apply(lambda x:(x//10)*10)).mean()['전력수요량(MWh)'].plot()
```

```python
df['풍향(deg)'].describe()
```

풍향이 어느정도 높을 때 높은 전력수요량을 보이나 degree는 360도와 0도가 같아 높은 풍향이나 낮은 풍향이 같으므로 명확한 상관관계를 나타낸다고 보기 어렵다.

### 시간과 전력수요량의 연관성

```python
plt.bar(df['월'].unique(),df.groupby('월').mean()['전력수요량(MWh)'])
```

여름, 겨울에 해당하는 7 ~ 8월과 12 ~ 2월에 전력수요량이 높은 것을 확인할 수 있다.

```python
plt.bar(df['일'].unique(),df.groupby('일').mean()['전력수요량(MWh)'])
```

일자가 전력수요량과 가지는 뚜렷한 상관관계를 확인할 수 없다.

```python
plt.bar(df['시간'].unique(),df.groupby('시간').mean()['전력수요량(MWh)'])
```

저녁시간대에 전력수요량이 가장 높았고 새벽 시간대에 전력수요량이 가장 낮았다.
