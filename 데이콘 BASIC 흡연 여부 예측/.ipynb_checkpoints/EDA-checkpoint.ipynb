## Import

```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['font.family'] ='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] =False
```

## 데이터 불러오기

```python
train = pd.read_csv('./train.csv')
train.head()
```

### 결측치 확인

```python
train.info()
```

결측치가 존재하지 않는 것을 확인할 수 있습니다.
### 기술 통계량 확인

```python
train.describe()
```

### Features 분포 그래프 확인

```python
train.drop(['ID','label'], axis=1).hist(figsize = (15,15))
plt.show()
```

시력, 공복혈당, 중성지방, 혈청 크레아티닌, 저밀도지단백, 요 단백, 간 효소율 등에서 정상 범주를 벗어나 비정상적으로 큰 수치를 확인할 수 있다.

```python
train.groupby('label').count()
```

4429명의 비흡연자와 2571명의 흡연자로 label이 63:37로 편향되었음을 확인할 수 있다.

### Features 상관관계 확인

```python
plt.figure(figsize = (15,15))
sns.heatmap(data = train.corr(), annot = True, cmap = 'Blues')
plt.show()
```

상관관계에서  키(cm)와 몸무게(kg), 몸무게(kg)와 BMI, 저밀도지단백과 콜레스테롤이 연관관계가 높은것을 확인할 수 있다.

체질량지수 BMI = 체중(kg) / 키(m)^2 으로 키와 몸무게에 직접적인 연관이 있는 지표이다.
저밀도지단백은 혈액을 통해 콜레스테롤을 운반하는 분자중 하나로 콜레스테롤과 직접적인 연관이 있는 지표이다.
