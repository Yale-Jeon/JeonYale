## Background analysis

주어진 데이터의 특성들이 가지는 의미를 알아보자.

MLM, HLM =  간 및 마우스의 간 대사효소와 화합물을 30분 동안 반응시킨 후, 대사되지 않고 남아있는 화합물의 양을(%) LC-MS/MS로 측정함으로써 화합물의 간 대사효소에 대한 안정성을 평가한 데이터

    AlogP	: logP는 극성용매와 비극성용매 사이의 물질분포를 측정한 값
    Molecular_Weight	: 분자량
    Num_H_Acceptors	: H 수용체의 수량
    Num_H_Donors	: H 도너의 수량
    Num_RotatableBonds	
    LogD	: 분자가 어떤 ph에서 어떤 형태로 존재하는지를 나타내는 지표
    Molecular_PolarSurfaceArea	: 분자의 극성 표면적의 합

MLM과 HLM의 경우 기술통계량 분석에서 maximum값이 130가량에 달하나 이론상 100%를 초과할 수는 없다. 용해도가 낮은 물질이 30분간 반응하면서 용해되어 화합물의 절대적인 량이 증가하면서 생긴 현상이며 이를 그대로 처리하거나 100%로 처리하거나 하는 두가지 방법을 사용 가능할 것으로 보인다.

### MorganFingerprint

Morgan Fingerprint를 사용하여 SMILES를 처리하자

```python
!pip install rdkit --user
```

```python
from rdkit import DataStructs
from rdkit.Chem import PandasTools, AllChem

PandasTools.AddMoleculeColumnToFrame(train_df,'SMILES','Molecule')
PandasTools.AddMoleculeColumnToFrame(test_df,'SMILES','Molecule')

def mol2fp(mol):
    fp = AllChem.GetHashedMorganFingerprint(mol, 6, nBits=4096)
    ar = np.zeros((1,), dtype=np.int8)
    DataStructs.ConvertToNumpyArray(fp, ar)
    return ar

# FPs column 추가
train_df["FPs"] = train_df.Molecule.apply(mol2fp)
test_df["FPs"] = test_df.Molecule.apply(mol2fp)

train_df
```

```python
len(train_df['FPs'][0])
```

MorganFingerprint는 Chemical들을 Morgan Algorithms으로 분석해 지문과 같은 고유의 형태로 변형하여 dataframe에 4096 길이를 가지는 numpy array로 저장한다.

'Molecule' column은 사용하지 않을테니 이대로 dataframe을 저장한다.
